The Tribonacci sequence Tn is defined as follows: 

T0 = 0, T1 = 1, T2 = 1, and Tn+3 = Tn + Tn+1 + Tn+2 for n >= 0.

Given n, return the value of Tn.
```
Input: n = 4
Output: 4
Explanation:
T_3 = 0 + 1 + 1 = 2
T_4 = 1 + 1 + 2 = 4
```

````python
def F_iter(n):
    """
    F_iter: the iteration function for Tribonacci series
    :type n: int
    :rtype: int
    Examples:
    
    F_iter(0) -> 0
    F_iter(1) -> 1
    F_iter(2) -> 1
    F_iter(3) -> 2
    F_iter(4) -> 4
    """
    if (n==0):
        return 0
    elif (n==1):
        return 1
    elif (n==2):
        return 1
    else:
        return F_iter(n-1)+F_iter(n-2) + F_iter(n-3)
    
````

We will next use class member function to test the same function code above

````python
class Solution(object):
    
    stored_list = [] # a list to store calculated Tribonacci numbers, e.g. [0,1,1,2,4,7,13]
    stored_dictionary = {} # a dictionary to store calculated Tribonacci numbers
    
    def trib(self, n):
        """
        trib: the iteration function for Tribonacci series
        :type n: int
        :rtype: int
        Examples:

        trib(0) -> 0
        trib(1) -> 1
        trib(2) -> 1
        trib(3) -> 2
        trib(4) -> 4
        """
        if (n==0):
            return 0
        elif (n==1):
            return 1
        elif (n==2):
            return 1
        else:
            return self.trib(n-1)+self.trib(n-2) + self.trib(n-3)
        
    def tribV2(self,n):
        """
        tribV2: using DP for Tribonacci series with List storage
        :type n: int
        :rtype: int
        """
        # check if the value is already in the list
        if len(self.stored_list) > n:
            return self.stored_list[n]
        # if not, expand the list until you have your answer
        else:
            self.extend_list(n)
            return self.stored_list[n]
            
    def tribV3(self,n):
        """
        tribV3: using DP for Tribonacci series with Dictionary storage
        Benefit of using Dictionary over List, is that it is 6.6x faster because dictionary uses
        hash lookup while list uses iteration for lookup.
        :type n: int
        :rtype: int
        """
        # check if the value is already in the dictionary
        if len(self.stored_dictionary) > n:
            return self.stored_dictionary[str(n)]
        # if not, expand the dictionary until you have your answer
        else:
            self.extend_dictionary(n)
            return self.stored_dictionary[str(n)]
        
    def tribV4(self,n):
        """
        tribV4: using DP for Tribonacci series with Dictionary storage, and precomputing all 38
        Tribonacci numbers required by the Leetcode checker.
        Benefit of using Dictionary over List, is that it is 6.6x faster because dictionary uses
        hash lookup while list uses iteration for lookup.
        :type n: int
        :rtype: int
        """
        # compute the stored_dictionary once and for all
        nMax = 38
        self.fill_dictionary(nMax)
        # lookup the dictionary using n
        return self.stored_dictionary[str(n)]

            
    def extend_list(self,n):
        """
        extend_list: given n, extend self.stored_list to have n+1 elements (0,1,...,n)
        :type n: int
        :rtype: none
        """
        # look at how many elements the self.stored_list already has, and start from there
        l0 = len(self.stored_list)
        if l0 < 3: # initialize stored_list as [0,1,1] if stored_list has less than 3 elements
            self.stored_list = [0,1,1]
            self.extend_list(n)
        elif n>=3:
            for i in range(l0,n+1):
                self.stored_list.append(self.stored_list[i-1]+self.stored_list[i-2]+self.stored_list[i-3])
                
    def extend_dictionary(self,n):
        """
        extend_dictionary: given n, extend self.stored_dictionary to have n+1 elements (0,1,...,n)
        note: do not just use integers as dictionary keys, they are illegal because Python identifiers cannot 
        start with a digit. We use the string version of the integers as keys here.
        :type n: int
        :rtype: none
        """
        # look at how many elements the self.stored_dictionary already has, and start from there
        l0 = len(self.stored_dictionary)
        if l0 < 3: # initialize stored_dictionary as {0:0,1:1,2:1} if stored_list has less than 3 elements
            self.stored_dictionary = {'0':0,'1':1,'2':1}
            self.extend_dictionary(n)
        elif n>=3:
            for i in range(l0,n+1):
                self.stored_dictionary[str(i)] = self.stored_dictionary[str(i-1)] \
                    +self.stored_dictionary[str(i-2)]+self.stored_dictionary[str(i-3)]
        
    def fill_dictionary(self,nMax):
        """
        fill_dictionary: given nMax, build the self.stored_dictionary to have nMax elements (0,1,...,nMax)
        :type n: int
        :rtype: none
        """
        # autofill the first three elements
        self.stored_dictionary = {'0':0,'1':1,'2':1}
        # fill the rest using the recursive relation
        for i in range(3,nMax+1):
            self.stored_dictionary[str(i)] = self.stored_dictionary[str(i-1)] \
                +self.stored_dictionary[str(i-2)]+self.stored_dictionary[str(i-3)]
````
