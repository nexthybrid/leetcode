740. Delete and Earn

You are given an integer array nums. You want to maximize the number of points you get by performing the following operation any number of times:

- Pick any nums[i] and delete it to earn nums[i] points. Afterwards, you must delete every element equal to nums[i] - 1 and every element equal to nums[i] + 1.

Return the maximum number of points you can earn by applying the above operation some number of times.

```
Example 1
Input: nums = [3,4,2]
Output: 6
Explanation: You can perform the following operations:
- Delete 4 to earn 4 points. Consequently, 3 is also deleted. nums = [2].
- Delete 2 to earn 2 points. nums = [].
You earn a total of 6 points.
```

````python
class Solution(object):
    
    # GLOBAL VARIABLES GOES HERE:
    # global variables don't even vanish after you destroy the class instance!
    # so the best practice is to NOT use them unless you know what you're doing
    # R = [] # a list to store R(n), the maximum reward of using the elements between
    # # 1 and n in the original nums list, i.e., R=[reward(1),reward(2),...]
    # D = {}
    
    def __init__(self):
        self.D = {}
    
    def deleteAndEarn(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # step 1: convert the list into a dictionary of total rewards
        self.D = {} # clear the self.D content before each problem
        self.D = self.list_to_dict(nums)
        nMax = max(self.D.keys())
        # step 2: use top-down approach in DP to calculate R(nMax)
        # initialize R[0],R[1],R[2]      
        self.R = [None] * nMax # initialize with None's
        self.R[0] = self.sum_x(1) # take all 1's (if they exist)
        if nMax > 1:
            self.R[1] = max(self.sum_x(1),self.sum_x(2))  # take all 1's or take all 2's (if they exist)
            if nMax > 2:
                self.R[2] = max(self.sum_x(1)+self.sum_x(3),self.sum_x(2))  # either take all 1's and 3's, or take all 2's (if they exist)
        return self.reward(nMax)
    
    def sum_x(self,x):
        """
        return the reward of taking an integer x from the original nums list,
        if x does not exist in nums, return zero
        :type x: int
        :rtype: int
        """
        if x in self.D.keys():
            return self.D[x]
        else:
            return 0
    
    def list_to_dict(self, nums):
        """
        Converting a list of integers into a dictionary. The keys are the integer, 
        the values are the total reward of taking that integer.
        YES, you can actually use integers as keys in dictionaries! I was wrong.
        :type nums: List[int]
        :rtype: dict
        """
        
        for num in nums:
            if num in self.D.keys():
                self.D[num] += num
            else:
                self.D[num] = num
        return self.D
    
    def reward(self,k):
        """
        reward(k) is the maximum reward of using the elements between 1 and k in the
        original nums list
        :type k: int
        :rtype: int
        """
        if self.R[k-1] != None or k <= 3:
            return self.R[k-1]
        else:
            if k in self.D.keys():
                if (k-1) in self.D.keys():
                    # store into R, and return
                    self.R[k-1] = max(self.D[k]+self.reward(k-2),self.D[k-1]+self.reward(k-3))
                    return self.R[k-1]
                else:
                    # store into R, and return
                    self.R[k-1] = self.D[k]+self.reward(k-2)
                    return self.R[k-1]
            else:
                return self.reward(k-1)
        
        
````
